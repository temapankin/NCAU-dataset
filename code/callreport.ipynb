```python
import pandas as pd

def combine_address(row):
    parts = [
        row['PhysicalAddressLine1'],
        row['PhysicalAddressLine2'] if pd.notnull(row['PhysicalAddressLine2']) and row['PhysicalAddressLine2'].strip() != '' else '',
        row['PhysicalAddressCity'],
        row['PhysicalAddressStateCode'],
        str(row['PhysicalAddressPostalCode']),
        row['PhysicalAddressCountry']
    ]
    return ', '.join([str(part).strip() for part in parts if part and str(part).strip() != ''])

# Create new column
locations_f['FullPhysicalAddress'] = locations_f.apply(combine_address, axis=1)
```

```python
# fill na values with 0 in MemberMinorityStatus_x column
merged_fs220['MemberMinorityStatus_x'].fillna(0, inplace=True)

(merged_fs220['MemberMinorityStatus_x'] == merged_fs220['MemberMinorityStatus_y']).all()
```

```python
merged_fs220['EligibleMinorityStatus_x'].fillna(0, inplace=True)

(merged_fs220['EligibleMinorityStatus_y'] == merged_fs220['EligibleMinorityStatus_x']).all()
```

```python
merged_fs220.drop(columns=['MemberMinorityStatus_y', 'EligibleMinorityStatus_y'], inplace=True)
merged_fs220.rename(columns={
    'MemberMinorityStatus_x': 'MemberMinorityStatus',
    'EligibleMinorityStatus_x': 'EligibleMinorityStatus'
}, inplace=True)
```

```python
extra_cols = ['CU_NUMBER', 'CYCLE_DATE', 'JOIN_NUMBER']

# Combine keys from the rename dictionary and the extra columns
keep_cols = extra_cols + list(column_map.keys())

# Filter the DataFrame to only these columns that actually exist in merged_upper
existing_cols = [col for col in keep_cols if col in merged_upper.columns]
merged_upper = merged_upper[existing_cols].copy()

# Rename the columns using the dictionary
df_filtered = merged_upper.rename(columns=column_map)
```

```python
df_filtered.columns
```

```python
df_filtered.head(3)
```

```python
locations_g
```

```python
df_merged = pd.merge(
    locations_g, 
    df_filtered, 
    on='CU_NUMBER',  # or whatever your common column is
    how='outer',     # keeps all rows from both DataFrames
    suffixes=('', '_dup')  # avoids conflict; duplicate columns get a suffix
)
```

```python
for col in df_merged.columns:
    if col.endswith('_dup'):
        original_col = col.replace('_dup', '')
        if original_col in df_merged.columns and df_merged[original_col].equals(df_merged[col]):
            df_merged.drop(columns=[col], inplace=True)
```
